```python
import pandas as pd
import numpy as np
df=pd.read_csv("testdata.csv")
df
```

```python
def get_weighted_score (food,service,ambience,price,scenario):
    mode=[[],[],[],[],[],[],[],[]]
    #       food,service,ambience,price
    mode[0]=[3.25,0.25,0.25,0.25] # = [0.8125,0.0625,0.0625,0.0625] for score of 1 (divide all by 4)
    mode[1]=[1,2.5,0.25,0.25]
    mode[2]=[1,0.25,2.5,0.25]
    mode[3]=[1,0.25,0.25,2.5]    
    mode[4]=[0.75,1.5,1.5,0.25]
    mode[5]=[0.75,1.5,0.25,1.5]
    mode[6]=[0.75,0.25,1.5,1.5]
    mode[7]=[1,1,1,1]
    
    score = (food*mode[scenario][0] + service*mode[scenario][1] + ambience*mode[scenario][2] + price*mode[scenario][3])/4
    
    return score
```

```python
df['f-score']=np.nan
df['s-score']=np.nan
df['a-score']=np.nan
df['p-score']=np.nan
df['as-score']=np.nan
df['sp-score']=np.nan
df['ap-score']=np.nan
df['asp-score']=np.nan

df
```

```python
for x in enumerate(df['id']):
    food = df['rest_food_rating'][x[0]]
    service = df['rest_srvc_rating'][x[0]]
    ambience = df['rest_ambi_rating'][x[0]]
    price = df['rest_prce_rating'][x[0]]
    
    df['f-score'][x[0]]=get_weighted_score(food,service,ambience,price,0)
    df['s-score'][x[0]]=get_weighted_score(food,service,ambience,price,1)
    df['a-score'][x[0]]=get_weighted_score(food,service,ambience,price,2)
    df['p-score'][x[0]]=get_weighted_score(food,service,ambience,price,3)
    df['as-score'][x[0]]=get_weighted_score(food,service,ambience,price,4)
    df['sp-score'][x[0]]=get_weighted_score(food,service,ambience,price,5)
    df['ap-score'][x[0]]=get_weighted_score(food,service,ambience,price,6)
    df['asp-score'][x[0]]=get_weighted_score(food,service,ambience,price,7)

df
```
